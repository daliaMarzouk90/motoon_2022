## Importing matplot

```python
import matplotlib.pyplot as plt
```

## First Example
create two plots sin and cos signal

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

## More Management for plots
use: fig = plt.figure() </br>
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html

```python
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

## Subplots

If you want to separate every plot command in an independent figure use the idea of subplots

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(3, 2, 6) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
ax = plt.subplot(3, 2, 1)
plt.plot(x, np.cos(x));
```

## Control Colors and Styles

```python
plt.plot(x, np.sin(x), color='green', linestyle='--') 
```

## Manage Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 2);
```

```python
from matplotlib.pyplot import xlim


plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

plt.title('graph')
plt.xlim(-2,10)
plt.ylim(-2,2)
plt.legend()
```

## Different graph types

```python
#1- bar plot
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.bar(x,y)

plt.subplot(1, 2, 2)
plt.barh(x,y)

plt.show()
```

```python
#2- pi plot
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
#3- histogram
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```
