### Class
<ul>
<li>Python is an object oriented programming (OOP) language.</li>
<li>The basic unit in OOP is the class </li>
<li>class simulates the real world objects</li>
<li>A class contains member variables which determin its properties and functions that determin how it behaves</li>
<li>You can define as much objects as you want from a certain class</li>
</ul>

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def print_info(self):
        print("Hello my name is " + self.name)
        print("And my age is " + self.age)

    def set_name(self, name):
        self.name = name

    def __accumelate_age(self):
        self.age += 1

    def ayear_passed(self):
        self.__accumelate_age()

p1 = Person("John", 36)
p1.print_info()
```
